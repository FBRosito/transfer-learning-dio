```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Image read with the function imread()

img = cv2.imread('lena_color.tiff')
width = img.shape[1]
print('Width in pixels: ', end='')
print(width)
height = img.shape[0]
print('Height in pixels: ', end='')
print(height)
print('Quantity of channels: ', end='')
print(img.shape[2])
```

```python
# To reduce or increase the size of the image, there is a ready-made OpenCV function,
# this is the ‘resize’ function.

# Note that you need to calculate the proportion of the height in relation to the
# width of the new image, otherwise it may be distorted

# Resize command
proportion = float(height/width)
new_width = 200 # in pixels
new_height = int(new_width*proportion)
new_size = (new_width, new_height)

resized_img = cv2.resize(img, new_size, interpolation = cv2.INTER_AREA)

# The ‘resize’ function uses a property defined here as cv2.INTER_AREA which
# is a specification of the mathematical calculation to resize the image.
# However, if the image is resized to a larger size, it is necessary
# to consider that there will be a loss of quality.
print("Resized image:")
cv2_imshow(resized_img)
```

```python
# Crop command

img = cv2.imread('lena_color.tiff')
clipping = img[50:300, 50:300]
print("Cropped image:")
cv2_imshow(clipping)

cv2.imwrite("clipping.png", clipping) # Save in disk
```

```python
# getRotationMatrix2D command

img = cv2.imread('lena_color.tiff')
(alt, lar) = img.shape[:2] # Capture width and heigh altura e largura

center = (lar // 2, alt // 2) # Find center
M = cv2.getRotationMatrix2D(center, 60, 1.0) # 60 degree rotation
rotaded_img = cv2.warpAffine(img, M, (lar, alt))

print("Image rotated by 60 degrees")
cv2_imshow(rotaded_img)

cv2.imwrite("rotaded.png", rotaded_img) # Save in disk
```

```python
# Image read in grayscale with the function imread()

img = cv2.imread('lena_color.tiff', cv2.IMREAD_GRAYSCALE)

print("Grayscale Image:")
cv2_imshow(img)

cv2.imwrite("grayscale.png", img) # Save in disk

# Apply binarization with a threshold value of 127
ret, binary_img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

print("\nBinary Image:")
cv2_imshow(binary_img)

cv2.imwrite("binary.png", binary_img) # Save in disk
```
